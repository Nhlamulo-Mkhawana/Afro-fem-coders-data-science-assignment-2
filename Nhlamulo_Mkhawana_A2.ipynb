```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
#Reading the csv file
Airbnb_NYC_2019 = pd.read_csv('Airbnb_NYC_2019.csv')
Airbnb_NYC_2019
```

```python
#Select a subset of the Airbnb NYC 2019 dataset — the subset should only contain the data where the neighbourhood group is Manhattan.
Manhattan_listing = Airbnb_NYC_2019.loc[["Manhattan" in title for title in Airbnb_NYC_2019 ["neighbourhood_group"]],:]
Manhattan_listing
```

```python
#How many rows are in the subset?
len(Manhattan_listing)
```

```python
#How many columns are in the subset?
len(Manhattan_listing.columns)
```

```python
#Create a histogram to visualize the distribution of prices for the Manhattan listings. 
plt.hist(Manhattan_listing ['price'],bins = np.arange(0, 1100, 40))
plt.title("Distribution of prices for Manhattan_listing in 2010")
plt.xlabel("Price (in US Dollars)")
plt.ylabel("Count")
plt.show()
```

```python
#Write a sentence to describe what you see
The distribution of prices has the highest count between 0-200 US dollars.
```

```python
#Create a scatter plot to visualize the number of reviews over the price for the Manhattan listings
plt.scatter('price', 'number_of_reviews', data= Manhattan_listing)
plt.title ("The number of reviews over the price for the Manhattan listings")
plt.xlabel("Price (in US Dollars)")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
#Write a sentence to describe what you see
The number of reviews decreases as the price increases.
```

```python
#Identify the “top 10” neighbourhoods from the Manhattan listings — the neighborhoods with the 10 highest counts i.e. number of listings. 
Top_10_Manhattan_listings = Manhattan_listing.groupby('neighbourhood')['neighbourhood'].count().nlargest(10)
Top_10_Manhattan_listings
```

```python
#Select the subset of data corresponding to the "top 10"
Top_10_listing=pd.merge(Manhattan_listing, Top_10_Manhattan_listings, left_on='neighbourhood', right_index=True)
Top_10_listing
```

```python
#How many rows are in the subset?
len(Top_10_listing)
```

```python
#How many columns are in the subset?
len(Top_10_listing.columns)
```
